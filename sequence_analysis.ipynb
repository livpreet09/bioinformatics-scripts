```python
# ============================================
# Bioinformatics Sequence Analysis Toolkit
# Author: Livpreet Kaur
# Description: DNA sequence analysis functions
# ============================================

# ---- 1. Count nucleotides ----
def count_nucleotides(sequence):
    """Count each nucleotide in a DNA sequence"""
    counts = {"A": 0, "T": 0, "G": 0, "C": 0}
    for base in sequence:
        if base in counts:
            counts[base] += 1
    return counts

# ---- 2. Transcribe DNA to RNA ----
def transcribe(dna):
    """Convert a DNA sequence to RNA"""
    return dna.replace("T", "U")

# ---- 3. Reverse complement ----
def reverse_complement(dna):
    """Return the reverse complement of a DNA sequence"""
    complement_map = {"A": "T", "T": "A", "G": "C", "C": "G"}
    complement = ""
    for base in dna:
        complement += complement_map[base]
    return complement[::-1]

# ---- 4. GC content ----
def gc_content(sequence):
    """Calculate GC content percentage of a sequence"""
    g = sequence.count("G")
    c = sequence.count("C")
    return (g + c) / len(sequence) * 100

# ---- 5. Find motif ----
def find_motif(dna, motif):
    """Find all positions of a motif in a DNA sequence (1-based)"""
    positions = []
    for i in range(len(dna)):
        if dna[i : i + len(motif)] == motif:
            positions.append(i + 1)
    return positions

# ---- 6. Parse FASTA file ----
def parse_fasta(filename):
    """Parse a FASTA file and return a dictionary of sequences"""
    sequences = {}
    current_name = ""
    with open(filename, "r") as f:
        for line in f:
            line = line.strip()
            if line.startswith(">"):
                current_name = line[1:]
                sequences[current_name] = ""
            else:
                sequences[current_name] += line
    return sequences

# ============================================
# TEST YOUR FUNCTIONS
# ============================================
dna = "GATATATGCATATACTT"

print("=== Nucleotide Counts ===")
print(count_nucleotides(dna))

print("\n=== RNA Transcription ===")
print(transcribe(dna))

print("\n=== Reverse Complement ===")
print(reverse_complement(dna))

print("\n=== GC Content ===")
print(f"{gc_content(dna):.2f}%")

print("\n=== Motif Search (ATAT) ===")
print(find_motif(dna, "ATAT"))
```
